```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def h00(t):
    return (1 + 2*t)*(1 - t)*(1 - t)

def h10(t):
    return t*(1 - t)*(1 - t)

def h01(t):
    return t*t*(3 - 2*t)

def h11(t):
    return t*t*(t-1)
```

```python
def Bi(x):
    if 0 <= x < 1:
        t = x - 0
        return 0.5*h11(t)
    elif 1 <= x < 2:
        t = x - 1
        return 0.5*h10(t) + h01(t)
    elif 2 <= x < 3:
        t = x - 2
        return h00(t) - 0.5*h11(t)
    elif 3 <= x < 4:
        t = x - 3
        return -0.5*h10(t)
    
    return 0.
```

```python
X = np.linspace(-1, 5, 6001)
y = np.array([Bi(x) for x in X])

plt.figure(figsize=(8,6))
plt.plot(X, y)
plt.tight_layout()
plt.savefig("Bi.png", dpi=100)
```
